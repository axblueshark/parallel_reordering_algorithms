```python
from petsc4py import PETSc
from scipy.io import mmread
```

### Converting matrices from .mtx to .bin

```python
mat_input_path = "../../matrices/mtx/"
mat_output_path = "../../matrices/bin/"
mat_name = "01_s_ex2"

# load .mtx file
A = mmread(mat_input_path + mat_name + ".mtx").tocsr()

# create PETSc matrix (AIJ = compressed sparse row)
petsc_A = PETSc.Mat().createAIJ(size=A.shape,
                                csr=(A.indptr, A.indices, A.data))

# write to PETSc binary format
viewer = PETSc.Viewer().createBinary(mat_output_path + mat_name + ".bin", PETSc.Viewer.Mode.WRITE)
petsc_A.view(viewer)
viewer.destroy()
```

### Converting vectors from .mtx to .bin

```python
vec_input_path = "../../matrices/mtx/vec/"
vec_output_path = "../../matrices/bin/vec/"
vec_name = "04_s_nasa4704_b"

# load .mtx file (column RHS vector)
b = mmread(vec_input_path + vec_name + ".mtx")

if hasattr(b, "toarray"):   # sparse matrix case
    b = b.toarray().ravel()
else:                       # already numpy.ndarray
    b = b.ravel()

# create PETSc vector
petsc_b = PETSc.Vec().createSeq(len(b))
petsc_b.setValues(range(len(b)), b)

# write to PETSc binary format
viewer = PETSc.Viewer().createBinary(vec_output_path + vec_name + ".bin", PETSc.Viewer.Mode.WRITE)
petsc_b.view(viewer)
viewer.destroy()
```
