```python
from petsc4py import PETSc
from scipy.io import mmread

mat_input_path = "../../matrices/mtx/"
mat_output_path = "../../matrices/bin/"
mat_name = "01_s_ex2"

# load Matrix Market file
A = mmread(mat_input_path + mat_name + ".mtx").tocsr()

# create PETSc matrix (AIJ = compressed sparse row)
petsc_A = PETSc.Mat().createAIJ(size=A.shape,
                                csr=(A.indptr, A.indices, A.data))

# write to PETSc binary format
viewer = PETSc.Viewer().createBinary(mat_output_path + mat_name + ".bin", PETSc.Viewer.Mode.WRITE)
petsc_A.view(viewer)
viewer.destroy()
```
