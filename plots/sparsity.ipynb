### Sparsity structure of both original and reordered matrix

```python
import os
import matplotlib.pyplot as plt

def draw_suitesparse_style(Asp, mat_name, k=200000, markersize=0.5):
    os.makedirs("./sparsity_structure", exist_ok=True)

    Aplot = Asp[:k, :k]

    fig, ax = plt.subplots(figsize=(8, 8))

    # blue structure
    ax.spy(Aplot, markersize=markersize, color="#1f77b4")

    # remove numbers/ticks, keep the frame
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel("")
    ax.set_ylabel("")
    ax.set_title("")

    # keep black rectangle
    for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_color("black")
        spine.set_linewidth(1.2)

    plt.tight_layout(pad=0.2)

    out = f"./sparsity_structure/{mat_name}_zoom.png"
    plt.savefig(out, dpi=600, bbox_inches="tight", pad_inches=0.02)
    print("Saved:", out)
    plt.show()
```

```python
mat_path = "../matrices/bin/Coarse_Matrix_15813251.bin"
Asp = load_matrix_for_plot(mat_path)

draw_suitesparse_style(Asp, "Coarse_Matrix_15813251", k=200000, markersize=0.5)
```
