# Scripts to obtain plots

```python
from petsc4py import PETSc
import matplotlib.pyplot as plt
from scipy.sparse import csr_matrix
from scipy.io import mmread
import numpy as np
```

## Sparsity structure

### Sparsity structure of both original and reordered matrix

```python
def draw_sparsity_structure(path, mat_name, title):
    # load matrix
    viewer = PETSc.Viewer().createBinary(path, "r")
    A = PETSc.Mat().create()
    A.load(viewer)
    viewer.destroy()

    # convert to SciPy CSR
    Ai, Aj, Av = A.getValuesCSR()
    nrows, ncols = A.getSize()
    Asp = csr_matrix((Av, Aj, Ai), shape=(nrows, ncols))

    # plot
    plt.figure(figsize=(25, 25))
    plt.spy(Asp, markersize=0.5)
    plt.title(title)

    ax = plt.gca()
    ax.ticklabel_format(axis='both', style='plain', useOffset=False)

    plt.savefig(f"./sparsity_structure/{mat_name}.png", dpi=300, bbox_inches="tight")
    plt.close()
```

```python
path_perm = "../matrices/bin/permuted/"
path_orig = "../matrices/bin/"
mat = "rajat21.bin"
#mat_perm = "great-britain_osm_permuted.bin"

draw_sparsity_structure(path_orig + mat, mat, "Original")

#draw_sparsity_structure(path_perm + mat_perm, "METISND" + mat_perm, "Permued")

plt.show()
```
