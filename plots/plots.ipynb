```python
import matplotlib.pyplot as plt
from scipy.io import mmread
import numpy as np
```

## Matrix structure

```python
def draw_matrix_structure(mat_name: str):
    A = np.array(mmread(f"../matrices/mtx/{mat_name}.mtx").todense())

    # mask exact zeros
    A_masked = np.ma.masked_equal(A, 0)

    # plot
    plt.figure(figsize=(8, 8))
    cmap = plt.get_cmap("seismic").copy()
    cmap.set_bad(color='white')  # show masked (zero) as white

    plt.imshow(A_masked, cmap=cmap, interpolation='nearest')
    # plt.colorbar(label="Matrix values")
    plt.savefig(f"./matrix_structure/mat_structure_{mat_name}.png", dpi=300, bbox_inches="tight")
    plt.close()
```

```python
draw_matrix_structure("01_s_bcsstk03")
draw_matrix_structure("02_u_saylr1")
```

## Sparsity structure

```python
def draw_sparsity_structure(mat_name: str):
    # load matrix (in mtx format)
    A = mmread(f"../matrices/mtx/{mat_name}.mtx")

    # plot sparsity structure
    plt.figure(figsize=(8, 8))
    plt.imshow(A.toarray() != 0, cmap="Blues")
    # plt.title("Matrix Sparsity Pattern")
    plt.savefig(f"./sparsity_structure/sparsity_{mat_name}.png", dpi=300, bbox_inches='tight')
    plt.close()
```

```python
draw_sparsity_structure("01_s_bcsstk03")
draw_sparsity_structure("02_u_saylr1")
```
